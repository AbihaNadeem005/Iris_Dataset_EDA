# **Task 1**: Exploring and Visualizing the Iris Dataset


```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Objective**: To load, explore, and visualize the Iris dataset using pandas, seaborn, and matplotlib to understand data patterns and distributions.

```python
# Load Dataset
df = sns.load_dataset('iris')
```

# **Dataset Overview**: We will now check the shape, column names, and first few rows of the dataset.

```python
# Data Inspection

# Shape of dataset
print(df.shape)
# Column name
print(df.columns)
# First five rows
df.head()
```

## **Summary Statistics**: We use .info() and .describe() to understand data types and statistical properties.

```python
# Summary Statistics

# Dataset information
df.info()
# Statistical summary
df.describe()
```

## **Scatter Plot**: This plot shows the relationship between sepal length and petal length across different species.

```python
# Data Visualization

# Scatter Plot: Relationship between features
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species'
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

## **Histograms**: Histograms are used to understand the distribution of each numerical feature.

```python
# Histograms: Data distribution
df.hist(figsize=(10,6))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

## **Box Plots**: Box plots help detect outliers and show data spread.

```python
# Box Plot: Outlier detection
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Box Plot of Iris Features")
plt.xticks(rotation=45)
plt.show()
```

## **Conclusion**
- The Iris dataset contains 150 samples and 3 species: Setosa, Versicolor, and Virginica.
- It includes 4 numerical features related to sepal and petal dimensions.
- Petal length and petal width are the most important features for distinguishing between species.
- Setosa species is clearly separable from the others in the scatter plot.
- No missing values are present, making the dataset clean and ready for analysis.
- Visualizations such as scatter plots, histograms, and box plots helped understand feature relationships and distributions.
